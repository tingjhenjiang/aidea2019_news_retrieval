```python
import nltk
nltk.download('punkt')
from nltk import tokenize as nltktokenize
class SentenceBreaker:
  ##regular expression for tokenize document to sentences
  def __init__(self):
    self.__to_sents_rule = re.compile(u'([^\n]+)[\n]*')

    ## 中文字元
    self.__ch=u'[\u4E00-\u9fa5]'

    ## 英文字元
    self.__eng=u'[a-zA-Z0-9_ａ-ｚＡ-Ｚ０-９－＿]'

    ## 英文字元+符號
    self.__eng_sym=u'[a-zA-Z0-9_ａ-ｚＡ-Ｚ０-９－＿\\/,.&\-]'

    ## 其他字元
    self.__other=u'[^\w\u4E00-\u9fa5\s]'

    ## 
    self.__to_tokens_rule = re.compile(
      self.__ch +
      u"|"+ self.__eng+self.__eng_sym+u"*" +
      u"|"+ self.__other
      )

  def ToSentsFirstRun(self,doc):
    '''split a document into sentences'''
    doc = re.sub('([。！？\?])([^」”’])', r"\1\n\2", doc)
    doc = re.sub('(\.{6})([^」”’])', r"\1\n\2", doc)
    doc = re.sub('(\…{2})([^」”’])', r"\1\n\2", doc)
    doc = re.sub('([。！？\?][」”’])([^，。！？\?])', r'\1\n\2', doc)
    sents=[m.group(0).strip(" \t\r\n") for m in re.finditer(self.__to_sents_rule, doc)]
    return sents
  def ToSents(self,doc):
    doc = re.sub('([a-zA-Z0-9_ａ-ｚＡ-Ｚ０-９－＿]{4,}\.)([a-zA-Z0-9_ａ-ｚＡ-Ｚ０-９－＿]{4,})',r'\1 \2',doc)
    breakedInChiDoc = self.ToSentsFirstRun(doc) #nltktokenize.sent_tokenize(doc)
    newdoc = []
    for sentence in breakedInChiDoc:
      if re.search(self.__eng, sentence)==None:
        newdoc.append(sentence)
      else:
        newdoc.extend(nltktokenize.sent_tokenize(sentence))
    return newdoc

  def ToTokens(self,sent):
    '''split a sentence into Chinese chars or English words'''
    tokens = [data for data in self.__to_tokens_rule.findall(sent)]
    return tokens
sentencebreaker = SentenceBreaker()
sentencebreaker = sentencebreaker.ToSents
#display(sentencebreaker.ToSents(testtargetstr.replace('&#39;','\'').replace('audiojs.events.ready(function(){var as = audiojs.createAll();});',' ')))
```

```python
class MyLRSchedule(tf.keras.optimizers.schedules.LearningRateSchedule):
  def __init__(self,
        initial_learning_rate=5e-5,
        max_learning_rate=5e-3,
        globalSteps=20000,
        warmupSteps=1000,
        name='warmupcosinerestartdecay',
        **settings):
    super(MyLRSchedule, self).__init__()
    self.initial_learning_rate = initial_learning_rate
    self.max_learning_rate = max_learning_rate
    self.warmupSteps = warmupSteps
    self.warmupSteps_float = tf.cast(warmupSteps, tf.float32)
    self.globalSteps = globalSteps
    settings.setdefault('first_decay_steps', globalSteps*0.5)
    settings.setdefault('t_mul', 0.3333)
    settings.setdefault('m_mul', 0.5)
    settings.setdefault('alpha', 0.0)
    settings.setdefault('power', 1.0)
    self.settings = settings
    self.name = name
    self.CosineDecayRestartsScheduleFunc = tf.keras.optimizers.schedules.CosineDecayRestarts(
        initial_learning_rate=max_learning_rate, first_decay_steps=settings['first_decay_steps'], t_mul=settings['t_mul'], m_mul=settings['m_mul'], alpha=settings['alpha'],
        name='CosineDecayRestarts'
    )
    self.learningRateDecayScheduleFunc = lambda step: self.CosineDecayRestartsScheduleFunc( step-warmupSteps )
  def linearWarmupLearningRateFunc(self, step):
    #display("input of warmup are {}".format(step))
    #(self.initial_learning_rate+(self.max_learning_rate-self.initial_learning_rate)/self.warmupSteps*step)
    step_float = tf.cast(step, tf.float32)
    returnv = tf.math.subtract(self.max_learning_rate,self.initial_learning_rate)
    returnv = tf.math.divide(returnv,self.warmupSteps_float)
    returnv = tf.math.multiply(returnv,step_float)
    returnv = tf.math.add(returnv,self.initial_learning_rate)
    return returnv
  def __call__(self, step):
    with tf.name_scope(self.name or 'WarmUp') as name:
      step_float = tf.cast(step, tf.float32)
      warmup_steps_float = tf.cast(self.warmupSteps, tf.float32)
      warmup_percent_done = tf.math.divide(step_float, warmup_steps_float)
      warmup_learning_rate = tf.math.multiply(  self.max_learning_rate, tf.math.pow(warmup_percent_done, self.settings['power']))
      return_learning_rate = tf.cond(tf.math.less_equal(step_float,warmup_steps_float), lambda: warmup_learning_rate, lambda: self.learningRateDecayScheduleFunc(step_float), name=name)
      return return_learning_rate
  def get_config(self):
    config = {
        'initial_learning_rate':self.initial_learning_rate,
        'max_learning_rate':self.max_learning_rate,
        'globalSteps':self.globalSteps,
        'warmupSteps':self.warmupSteps,
        'name':self.name,
        'settings':self.settings,
    }
    #base_config = super(MyLRSchedule, self).get_config()
    return config

def remove_all_zeros_rows(srctensor):
  boolean_mask = tf.cast(tf.reduce_sum(tf.abs(srctensor), 1), dtype=tf.bool)
  no_zeros = tf.boolean_mask(srctensor, boolean_mask, axis=0)
  return no_zeros
```

```python
newsfullEmbeddingsDict_filename = "newsfullEmbeddingsDict.json"
daskapplyNewsEmbeddingMeta = ('newsEmbedding',object)

def findrowAndExtractEmbedding(location=''):
  try:
    assert location!='foo', 'location must not be foo'
    #print('location is {}'.format(location))
    if type(location)==str:
      #print('complete step prefix check')
      rowAtlocation = newsdf_for_processingEmbedding[newsdf_for_processingEmbedding['News_Index']==location]
    else:
      rowAtlocation = newsdf_for_processingEmbedding.iloc[[location],:]
    assert len(rowAtlocation.index)==1, 'number of rows in rowAtlocation must be 1'
    #print('complete step row check')
    need_sentences = rowAtlocation['sentencebreakNews_FullTitleContent'].iloc[0]
    needEmbedding = extract_globalavgpooled_embedding(need_sentences)
    print("completed {}".format(location))
  except Exception as embed_error:
    needEmbedding = "in findrowAndExtractEmbedding at {} for {}".format(location,embed_error)
  return needEmbedding

def mean_pooling(token_embeddings, attention_mask, lib='pt'):
  if lib=='tf':
    input_mask_expanded = tf.expand_dims(attention_mask, axis=-1)
    input_mask_expanded = tf.broadcast_to(input_mask_expanded, tf.shape(token_embeddings))
    input_mask_expanded = tf.cast(input_mask_expanded, dtype=tf.float32)
    sum_embeddings = tf.math.reduce_sum(token_embeddings * input_mask_expanded, axis=1)
    sum_mask = tf.math.reduce_sum(input_mask_expanded, axis=1)
    sum_mask = tf.clip_by_value(sum_mask, clip_value_min=1e-9, clip_value_max=9999^9)
  else:
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)
    sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
  returnres = sum_embeddings / sum_mask
  del input_mask_expanded,sum_embeddings,sum_mask
  return returnres

def extract_sentence_meanpooled_embeddings_from_transformer(sentences, transformer_model=ckipmodel):
  try:
    tokenizedRes = ckiptokenizer(sentences, **tokenizerSettingsDoc)
    assert 'input_ids' in tokenizedRes, 'input_ids must be in tokenizedRes but {} detected'.format(tokenizedRes)
    tokenizedRes = tokenizedRes.to(device) if device!='cpu' else tokenizedRes
    resEmbedding = transformer_model(**tokenizedRes)
    resEmbedding = resEmbedding.logits
    assert isinstance(resEmbedding.shape, tuple), 'logits shape must be a tuple but {} detected'.format(resEmbedding)
    resEmbedding = mean_pooling(resEmbedding, tokenizedRes['attention_mask'])
    assert isinstance(resEmbedding.shape, tuple), 'mean_pooling logits shape must be a tuple but {} detected'.format(resEmbedding)
    del tokenizedRes
    #print("complete extract meanpooled step 5")
    #embeddings_to_return_cpu = resEmbedding.cpu().detach().numpy()
    #print("complete extract meanpooled step 6")
    #del resEmbedding
    #print("complete extract meanpooled step 7")
    #embeddings_to_return = embeddings_to_return_cpu
  except Exception as e:
    resEmbedding = "in function extract_sentence_meanpooled_embeddings_from_transformer at {} for {}".format(sentences,e)
  return resEmbedding

def extract_globalavgpooled_embedding(need_sentences, transformer_model=ckipmodel):
  try:
    assert need_sentences!='foo', 'need_sentences must not be foo'
    assert type(need_sentences)==list, 'type of need_sentences must be list but {} for type {} detected'.format(need_sentences, type(need_sentences))
    #print('complete step news_sentence check')
    #needEmbedding = sentenceTransformerFromPackageModel.encode(need_sentences)
    needEmbedding = extract_sentence_meanpooled_embeddings_from_transformer(need_sentences, transformer_model=transformer_model)
    assert type(needEmbedding)!=str, 'needEmbedding must not be str but {} detected at location {}'.format(needEmbedding, need_sentences)
    assert needEmbedding.shape[1]>=512, 'shape1 must be >512 but {} detected'.format(needEmbedding)
    #print('complete step embedding type check, type is {}'.format(type(needEmbedding)))
    if str(type(needEmbedding))=="<class 'torch.Tensor'>":
      needEmbedding = torch.permute(needEmbedding, dims=(1,0) )
      needEmbedding = torch.nn.AdaptiveAvgPool1d(1)(needEmbedding)
      needEmbedding = torch.reshape(needEmbedding, (-1,))
      needEmbedding_cpu = needEmbedding.cpu().detach().numpy()
      del needEmbedding
      needEmbedding = needEmbedding_cpu
    else:
      needEmbedding = tf.expand_dims(needEmbedding, axis=0)
      needEmbedding = tf.keras.layers.GlobalAveragePooling1D()(needEmbedding)
      needEmbedding = tf.reshape(needEmbedding, -1)
      needEmbedding = needEmbedding.numpy()
    assert needEmbedding.shape[0]>=512, 'shape0 must be >=512 but {} detected, type {}'.format(needEmbedding, type(needEmbedding))
    #print('complete step tensor transform check')
    #needEmbedding = needEmbedding.tolist()
    #assert type(needEmbedding)==list, 'type of needEmbedding must be list but type {} detected'.format(need_sentences, type(needEmbedding))
    #print('complete step np-list transform check')
    #  print("successful at {} for {}".format(location, needEmbedding))
  except Exception as embed_error:
    needEmbedding = 'in extract_globalavgpooled_embedding {}'.format(embed_error)
  return needEmbedding

def readWritenewsfullEmbeddingsDict(mode='r', newsfullEmbeddingsDict={}):
  if mode=='w':
    newsfullEmbeddingsDict_file = open(newsfullEmbeddingsDict_filename, "w")
    newsfullEmbeddingsDict = json.dump(newsfullEmbeddingsDict, newsfullEmbeddingsDict_file)
  else:
    newsfullEmbeddingsDict_file = open(newsfullEmbeddingsDict_filename, "r")
    newsfullEmbeddingsDict = json.load(newsfullEmbeddingsDict_file)
  newsfullEmbeddingsDict_file.close()
  return newsfullEmbeddingsDict

def get_newsdf_for_processingEmbedding():
  t_newsdf_for_processingEmbedding = pd.merge(left=nc, right=newsdf, on="News_URL", how="left")
  t_newsdf_for_processingEmbedding = t_newsdf_for_processingEmbedding[~pd.isnull(t_newsdf_for_processingEmbedding['News_Index'])]
  t_newsdf_for_processingEmbedding = t_newsdf_for_processingEmbedding[~pd.isnull(t_newsdf_for_processingEmbedding['sentencebreakNews_FullTitleContent'])]
  t_newsdf_for_processingEmbedding = t_newsdf_for_processingEmbedding.drop_duplicates(subset=['News_Index'])
  t_newsdf_for_processingEmbedding = t_newsdf_for_processingEmbedding.loc[:,['News_Index','sentencebreakNews_FullTitleContent']]
  t_newsdf_for_processingEmbedding = t_newsdf_for_processingEmbedding.sort_values(by=['News_Index'])
  t_newsdf_for_processingEmbedding = t_newsdf_for_processingEmbedding.reset_index(drop=True)
  return t_newsdf_for_processingEmbedding
```

```python
import sentencepiece as spm
from prepro_utils import preprocess_text, encode_ids
from functools import reduce
def turn_list_into_zeros_of_equal_length(lst):
    if lst==None:
        return None
    else:
        return np.zeros_like(np.array(lst)).tolist()
def concentrate_list_by_seperator(x, y, sep=[sp_model_marks['<sep>']], seg_id=False, pad_with_incre=False, printvar='none'): #, flatappend=False
    if type(printvar)!=str:
        print(printvar)
    if seg_id==False:
        if y!=None:
            if sep!=None:
                newlst = [*x, *sep, *y]
            else:
                newlst = [*x, *y]
        else:
            newlst = [*x]
    else:
        try:
            seg_id_to_pad = x[-1]
            if pad_with_incre==True:
                seg_id_to_pad = seg_id_to_pad+1
            if y!=None:
                to_be_appended = [seg_id_to_pad]*len(y)
                if sep!=None:
                    newlst = [*x, *[x[-1]], *to_be_appended  ] 
                else:
                    newlst = [*x, *to_be_appended ] 
            else:
                newlst = [*x]
        except Exception as e:
            print("x is {} and y is {} and and error is {}".format(x, y, str(e) ))
            return 'it is none'
    return newlst
def seperate_sentences_and_encode_and_join(text, tokenizer=xlnet_tokenizer, seperator=[sp_model_marks['<eop>']], returnsegids=False, inputtype="str", splitter=[sp_model_marks['<eop>']], pad_with_incre=False): #, flatappend=False
    if inputtype=='str':
        sptext = preprocess_text(text, lower=False)
        sptext = utilized_seg_funcs.ToSents(sptext)
        ids = [tokenizer.encode(t) for t in sptext if not(t is None)]
    else:
        ids = list(split_at_values(text, splitter))
    #if len(ids)==0:
    #    print("error at text where text are")
    #    print(text)
    #    print(str(text))
    #    sys.exit()
    if len(ids)<=1:
        ids = [*ids, None]
    if returnsegids==True:
        ids = [turn_list_into_zeros_of_equal_length(id) for id in ids]
        ids_in_zeros_with_sep = reduce(lambda x, y: concentrate_list_by_seperator(x, y, sep=seperator, seg_id=True, pad_with_incre=pad_with_incre), ids) #, flatappend=flatappend
        return ids_in_zeros_with_sep
    else:
        #try:
        ids_with_sep = reduce(lambda x, y: concentrate_list_by_seperator(x, y, sep=seperator, seg_id=False, pad_with_incre=pad_with_incre), ids) #, flatappend=flatappend

        #except Exception as e:
        #    print("ids are")
        #    print(ids)
        #    print("error is {}".format(str(e)))
        #    print("x is ")
        #    print(x)
        #    print("y is ")
        #    print(y)
        return ids_with_sep
def padding_zeros_at_head(lst, paddingvalue=sp_model_marks['<pad>'], fulllen=seq_len-2):
    #padding value should be symbol of <pad> for token or 0 for seg id
    if len(lst)>=fulllen:
        lst = lst
    else:
        lst = [*( [paddingvalue]*int(fulllen-len(lst)) ), *lst]
    return lst
def add_symbol(lst, returnsegids=False, appendmarks=[sp_model_marks['<sep>'], sp_model_marks['<cls>']], tobeaddedpos="end"):
    if returnsegids==True:
        if tobeaddedpos=="end":
            lst = [*lst, *([lst[-1]]*len(appendmarks)) ] #segid時最尾部的element重複
        else:
            lst = [*([lst[-1]]*len(appendmarks)), *lst] #segid時最尾部的element重複
    else:
        if tobeaddedpos=="end":
            lst = [*lst, *appendmarks ] #加上符號
        else:
            lst = [*appendmarks, *lst] #加上符號
    return lst
def encode_for_input_seq(text, htype='token', tokenizer=xlnet_tokenizer, seperator=[sp_model_marks['<eop>']], appendmarks=[sp_model_marks['<sep>'], sp_model_marks['<cls>']], tobeaddedpos="end", for_debug_df=None):
    try:
        if htype=='token':
            seq = seperate_sentences_and_encode_and_join(text, tokenizer=tokenizer, seperator=seperator, returnsegids=False)
            if appendmarks!=None:
                seq = add_symbol(seq, appendmarks=appendmarks, tobeaddedpos=tobeaddedpos)
        else:
            seq = seperate_sentences_and_encode_and_join(text, tokenizer=tokenizer, seperator=seperator, returnsegids=True)
            if appendmarks!=None:
                seq = add_symbol(seq, returnsegids=True, appendmarks=appendmarks, tobeaddedpos=tobeaddedpos)
        return seq
    except Exception as e:
        print("encode_for_input_seq error reason is {}".format(str(e)))
        print("error at")
        if for_debug_df!=None:
            print(str(for_debug_df))
        sys.exit()
    #return [token_seq, segid_seq, mem_len]
def split_at_values(lst, values):
    indices = [i for i, x in enumerate(lst) if x in values] # 0 + i end point
    indices_plus_one = [i+1 for i in indices] #starting point
    for start, end in zip([0, *indices_plus_one], [*indices_plus_one, len(lst)]):
        if not(lst[start:end] is None):
            yield lst[start:end]
def shorten_doc_by_filter_sentences_from_querywords(doc_tokenized, querywords_in_list, query_tokenized, seq_len_limit=seq_len, eodmark=sp_model_marks['<eod>'], para_splitter_list=[sp_model_marks['<eop>']]):
    if len(query_tokenized)+1+len(doc_tokenized)+2<seq_len_limit:
        #doc_tokenized.append(sp_model_marks['<eod>'])
        return [*doc_tokenized, eodmark]
    else:
        if type(querywords_in_list)!=list:
            querywords_in_list = querywords_in_list.split(" ")
        #querywords_tokenized = [item for word in querywords_in_list for item in xlnet_tokenizer.encode(word)]
        querywords_tokenized = querywords_in_list
        doc_paragraphs = list(split_at_values(doc_tokenized, para_splitter_list))
        doc_paragraphs_with_key = { k:v for k,v in enumerate(doc_paragraphs) }
        doc_paragraphs_each_para_len = { k:len(v) for k,v in enumerate(doc_paragraphs) }
        doc_paragraphs_each_para_intersections = {k:len(set(querywords_tokenized) & set(v) - set(sp_model_stopwords_marks)) for k,v in doc_paragraphs_with_key.items()}
        doc_paragraphs_each_para_intersections = dict(sorted(doc_paragraphs_each_para_intersections.items(), key=lambda x: x[1], reverse=True))
        #{k:len(set(querywords_tokenized) & set(v) - set(sp_model_stopwords_marks)) for k,v in doc_paragraphs_with_key.items()}
        intersection_paras_keys = []#[i for i,x in enumerate(doc_paragraphs_each_para_intersections) if x>0]
        test_length = 0
        for para_key, para_v in doc_paragraphs_each_para_intersections.items():
            test_length = test_length+doc_paragraphs_each_para_len[para_key]
            if len(query_tokenized)+1+test_length+3>seq_len_limit:
                break
            else:
                intersection_paras_keys.append(para_key)
        intersection_paras_keys.sort()
        final_ret_doc = [item for key in intersection_paras_keys for item in doc_paragraphs_with_key[key]]
        return [*final_ret_doc, eodmark]
def filter_out_in_lst(srclst, matchlst):
    if matchlst==[] or matchlst==None:
        return srclst
    else:
        return list(filter(lambda x: not(x in matchlst), srclst))
```

* \<CLS\>：在做分類任務時其最後一層的 repr. 會被視為整個輸入序列的 repr.

* \<UNK\>：沒出現在 BERT 字典裡頭的字會被這個 token 取代
* \<PAD\>：zero padding 遮罩，將長度不一的輸入序列補齊方便做 batch 運算
* \<MASK\>：未知遮罩，僅在預訓練階段會用到

```python
from tensorflow.python.keras import backend as K

def binary_focal_loss(gamma=2., alpha=.25):
    """
    Binary form of focal loss.
      FL(p_t) = -alpha * (1 - p_t)**gamma * log(p_t)
      where p = sigmoid(x), p_t = p or 1 - p depending on if the label is 1 or 0, respectively.
    References:
        https://arxiv.org/pdf/1708.02002.pdf
    Usage:
     model.compile(loss=[binary_focal_loss(alpha=.25, gamma=2)], metrics=["accuracy"], optimizer=adam)
    """
    def binary_focal_loss_fixed(y_true, y_pred):
        """
        :param y_true: A tensor of the same shape as `y_pred`
        :param y_pred:  A tensor resulting from a sigmoid
        :return: Output tensor.
        """
        pt_1 = tf.where(tf.equal(y_true, 1), y_pred, tf.ones_like(y_pred))
        pt_0 = tf.where(tf.equal(y_true, 0), y_pred, tf.zeros_like(y_pred))

        epsilon = K.epsilon()
        # clip to prevent NaN's and Inf's
        pt_1 = K.clip(pt_1, epsilon, 1. - epsilon)
        pt_0 = K.clip(pt_0, epsilon, 1. - epsilon)

        return -K.sum(alpha * K.pow(1. - pt_1, gamma) * K.log(pt_1)) \
               -K.sum((1 - alpha) * K.pow(pt_0, gamma) * K.log(1. - pt_0))

    return binary_focal_loss_fixed


def categorical_focal_loss(gamma=2., alpha=.25):
    """
    Softmax version of focal loss.
           m
      FL = ∑  -alpha * (1 - p_o,c)^gamma * y_o,c * log(p_o,c)
          c=1
      where m = number of classes, c = class and o = observation
    Parameters:
      alpha -- the same as weighing factor in balanced cross entropy
      gamma -- focusing parameter for modulating factor (1-p)
    Default value:
      gamma -- 2.0 as mentioned in the paper
      alpha -- 0.25 as mentioned in the paper
    References:
        Official paper: https://arxiv.org/pdf/1708.02002.pdf
        https://www.tensorflow.org/api_docs/python/tf/keras/backend/categorical_crossentropy
    Usage:
     model.compile(loss=[categorical_focal_loss(alpha=.25, gamma=2)], metrics=["accuracy"], optimizer=adam)
    """
    def categorical_focal_loss_fixed(y_true, y_pred):
        """
        :param y_true: A tensor of the same shape as `y_pred`
        :param y_pred: A tensor resulting from a softmax
        :return: Output tensor.
        """

        # Scale predictions so that the class probas of each sample sum to 1
        y_pred /= K.sum(y_pred, axis=-1, keepdims=True)

        # Clip the prediction value to prevent NaN's and Inf's
        epsilon = K.epsilon()
        y_pred = K.clip(y_pred, epsilon, 1. - epsilon)

        # Calculate Cross Entropy
        cross_entropy = -y_true * K.log(y_pred)

        # Calculate Focal Loss
        loss = alpha * K.pow(1 - y_pred, gamma) * cross_entropy

        # Sum the losses in mini_batch
        return K.sum(loss, axis=1)

    return categorical_focal_loss_fixed
```
